# Time to calculate final revenue

```python
import pandas as pd
import numpy as np
```

```python
revenue_MF = pd.read_csv('~/Desktop/revenue_MF.csv')
revenue_CC = pd.read_csv('~/Desktop/revenue_CC.csv')
revenue_CL = pd.read_csv('~/Desktop/revenue_CL.csv')
```

```python
revenue_MF = revenue_MF.drop('Unnamed: 0', 1)
revenue_MF["Lable_MF"].replace({1: "MF"}, inplace=True)
revenue_MF.rename(columns={'Lable_MF': 'Lable'}, inplace=True)
revenue_MF.head()
```

```python
revenue_CC = revenue_CC.drop('Unnamed: 0', 1)
revenue_CC["Lable_CC"].replace({1: "CC"}, inplace=True)
revenue_CC.rename(columns={'Lable_CC': 'Lable'}, inplace=True)
revenue_CC.head()
```

```python
revenue_CL = revenue_CL.drop('Unnamed: 0', 1)
revenue_CL["Lable_CL"].replace({1: "CL"}, inplace=True)
revenue_CL.rename(columns={'Lable_CL': 'Lable'}, inplace=True)
revenue_CL.head()
```

```python
#stack vertically the results of the three columns
vertical_stack = pd.concat([revenue_MF, revenue_CC], axis=0, sort = False)
final = pd.concat([vertical_stack, revenue_CL], axis=0, sort = False)
```

```python
#sort by probability score
final = final.sort_values (by = 'Probability', ascending = False)

#we can contact each client only once, thus, it is important to drop duplicated clients
final = final.drop_duplicates(subset ="Client", 
                     keep = 'first')

final = final.reset_index ( drop = True)
```

```python
final_pivot = final.pivot_table (values = 'Pred_Revenue',
                          index = 'Lable',
                          aggfunc = 'sum')
```
